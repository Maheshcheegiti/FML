```python
#most of the utilities under the pyplot submodule
import matplotlib.pyplot as plt
import numpy as np
```

```python
# example to plot a graph
x = np.array([0,4])
y = np.array([0,8])
plt.plot(y)
plt.show()
```

```python
#marker and line
x=[1,4,8,12,16]
y=[8,1,9,6,2]
plt.plot(x,y,marker='P',ms=20,mfc='g',mec='r',ls=':',lw=5)
# marker markersize markerfacecolor markeredgecolor linestyle linewidth color 
#marker references likw o * . , x X + P s D d p H ...
plt.show()
```

```python
#sin graphs
x = np.linspace(0,2*3.14,100)
print(x)
y=np.sin(x)
plt.plot(y)
plt.show()
```

```python
#multiple lines
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1)
plt.plot(x2,y2)
plt.show()
```

```python
#label and Title
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Graph",fontdict={'family':'serif','color':'red','size':30},loc='right'); #loc
plt.plot(x, y)
plt.show()
```

```python
#grid
plt.plot(x,y)
plt.grid(axis='y') #axis='x','y','both' and we can add color linestyle linewidth
```

```python
#subplot
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1) #row coloumn index

#title

plt.plot(x,y)
plt.title("subplot 1")
#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.title("subplot 2")
#title

plt.plot(x,y)

#super title
plt.suptitle("Graph")
plt.show()
```

```python
#scatter plots
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y,color='r')
plt.show()
```

```python
#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y) #color

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)

plt.show()
```

```python
#each dot each color
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c=colors)

plt.show()
```

```python
#color Map :  a list of colors, where each color has a value that ranges from 0 to 100
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])  #and also sizes

plt.scatter(x, y, c=colors, cmap='Blues',alpha=1) #Accect, Blues and we can apply alpha

#color bar print
plt.colorbar()

plt.show()
```

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='Accent')

plt.colorbar()

plt.show()
```

```python
#bar graphs

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y) #color linewidth  bydefault = 0.8 lw
plt.show()
```

```python
#horizontal bargraphs
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x,y,color='r',height=0.3) #color lineheight 
plt.show()
```

```python
#Histograms : a graph showing frequency distributions.
x = np.random.normal(100,10,250)
plt.hist(x,color='red')
plt.show()
```

```python
#pie charts
y = np.array([35, 25, 25, 15])
#mylabels
mylables=["Apples","Bananans","Grapes","Cherries"]
plt.pie(y,labels=mylabels)  #labels startangle
plt.show() 
```

```python
#pie charts explode and shadow
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.5, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow="true") #colors = mycolors
plt.legend(title="Fruits",loc="right", bbox_to_anchor=(1.5, 1))
plt.show()
```

```python
#box-plot
x = [1,2,3,4,5,6,7,8,9]
plt.boxplot(x)
plt.show()
```
